## Criando uma tabela fato da distribuição de medicamentos


```python
# --- CONFIGURAÇÕES GERAIS ---
random.seed(42)
np.random.seed(42)

# Definição de Capitais para Regra de Negócio
CAPITAIS_NE = [
    "Recife", "Salvador", "Fortaleza", "São Luís", "Maceió",
    "João Pessoa", "Teresina", "Natal", "Aracaju"
]

# --- 1. CARREGAMENTO DAS TABELAS DIMENSÃO ---

try:
    # 1. Carregar Farmácias (108 lojas)
    #
    df_farmacias = pd.read_csv("dim_farmacias_final.csv")

    # 2. Carregar Lista Mestra de Medicamentos (3080 SKUs)
    # Ajuste o caminho conforme necessário
    df_produtos = pd.read_csv("lista_medicamentos_expandida.csv")

    print(f"Farmácias carregadas: {len(df_farmacias)}")
    print(f"Produtos carregados: {len(df_produtos)}")

except FileNotFoundError as e:
    print(f"ERRO CRÍTICO: Arquivo não encontrado - {e}")
    # Mock para não quebrar o código se você for testar sem arquivos
    df_farmacias = pd.DataFrame([{'cod_farmacia': 1, 'nome_farmacia': 'Teste', 'cidade': 'Recife'}])
    df_produtos = pd.DataFrame([{'cod_medicamento': 700000, 'nome_medicamento': 'Dipirona', 'categoria': 'Isentos', 'fabricante': 'EMS', 'dosagem': '500mg', 'apresentacao': 'Cp', 'substancia_principal': 'Dipirona'}])

# --- 2. GERAÇÃO DO DATASET DE ESTOQUE ---

print("Gerando estoque distribuído (pode levar alguns segundos)...")

dados_estoque = []

# garantir cobertura
for _, farmacia in df_farmacias.iterrows():

    eh_capital = farmacia['cidade'] in CAPITAIS_NE

    for _, produto in df_produtos.iterrows():

        categoria = produto['categoria']

        # --- LÓGICA DE QUANTIDADE BASE ---
        # Produtos de alta saída tem base maior
        if categoria in ["Higiene e Beleza", "Medicamentos Isentos", "Diversos"]:
            qtd_base = random.randint(20, 150)
        elif categoria == "Uso Contínuo":
            qtd_base = random.randint(15, 80)
        elif categoria == "Antibióticos e Controlados":
            qtd_base = random.randint(5, 40)
        elif categoria == "Refrigerados": # Vacinas/Insulinas
            qtd_base = random.randint(2, 15)
        else:
            qtd_base = random.randint(5, 20)

        # --- MULTIPLICADOR DE LOCALIZAÇÃO ---
        if eh_capital:
            # Capitais tem estoque 30% a 80% maior
            fator_loc = random.uniform(1.3, 1.8)
        else:
            # Interior tem estoque 20% a 50% menor
            fator_loc = random.uniform(0.5, 0.8)

        # --- REGRA DE RARIDADE / ESTOQUE ZERO ---
        # Medicamentos raros/caros podem não ter no interior
        qtd_final = int(qtd_base * fator_loc)

        if not eh_capital:
            # Se for interior, chance de não ter estoque de itens complexos
            if categoria in ["Refrigerados", "Antibióticos e Controlados"]:
                if random.random() < 0.15: # 15% de chance de ser zero
                    qtd_final = 0
            # Mesmo itens comuns podem faltar ocasionalmente
            elif random.random() < 0.02:
                qtd_final = 0

        # Montagem da linha
        row = {
            # Dados do Produto
            "cod_medicamento": produto["cod_medicamento"],
            "nome_medicamento": produto["nome_medicamento"],
            "substancia_principal": produto["substancia_principal"],
            "apresentacao": produto["apresentacao"],
            "dosagem": produto["dosagem"],
            "fabricante": produto["fabricante"],
            "categoria": produto["categoria"],

            # Dados de Estoque
            "quantidade_estoque": qtd_final,

            # Dados da Farmácia
            "cod_farmacia": farmacia["cod_farmacia"],
            "nome_farmacia": farmacia["nome_farmacia"],
            "cidade": farmacia["cidade"]
        }
        dados_estoque.append(row)

# Criar DataFrame
df = pd.DataFrame(dados_estoque)

# --- 3. INTRODUÇÃO DE IMPERFEIÇÕES ---

def introduzir_erro(valor, chance=0.03, tipo='texto'):
    if pd.isna(valor): return valor
    prob = random.random()
    if prob < chance:
        if tipo == 'nan': return np.nan
        if tipo == 'texto_upper': return str(valor).upper()
        if tipo == 'texto_lower': return str(valor).lower()
    return valor

# Aplicar imperfeições solicitadas
# A. Dados Faltantes (~2%)
cols_sujas_nan = ['nome_farmacia', 'substancia_principal', 'fabricante', 'apresentacao']
for col in cols_sujas_nan:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: introduzir_erro(x, chance=0.02, tipo='nan'))

# B. Sujeira na Cidade (~1% nulo, ~15% formatação)
df['cidade'] = df['cidade'].apply(lambda x: introduzir_erro(x, chance=0.01, tipo='nan'))
df['cidade'] = df['cidade'].apply(lambda x: introduzir_erro(x, chance=0.10, tipo='texto_upper'))
df['cidade'] = df['cidade'].apply(lambda x: introduzir_erro(x, chance=0.05, tipo='texto_lower'))

# C. Erros na Quantidade
def sujar_quantidade(val):
    prob = random.random()
    if prob < 0.001: return -val      # Valor negativo (raro)
    if prob < 0.001: return 99999     # Outlier (raro)
    return val

df['quantidade_estoque'] = df['quantidade_estoque'].apply(sujar_quantidade)

# --- 4. EXPORTAÇÃO ---
print(f"Dataset Final Gerado com {len(df)} registros.")
print(df[['nome_medicamento', 'cidade', 'quantidade_estoque']].head(10))

filename = "distribuicao_medicamentos_completa.csv"
df.to_csv(filename, index=False, encoding='utf-8')
print(f"Arquivo '{filename}' pronto para download.")
```

```python
# --- 1. CARREGAMENTO E PREPARAÇÃO ---

try:
    distrib = pd.read_csv("distribuicao_medicamentos_completa.csv")
    print(f"Base carregada: {len(distrib)} linhas.")
except FileNotFoundError:
    print("ERRO: Arquivo não encontrado. Verifique o caminho.")
    # Seção de emergência para o código não quebrar em caso de testar sem o arquivo
    distrib = pd.DataFrame({
        'cod_farmacia': [1001]*10, 'cod_medicamento': range(700000, 700010),
        'quantidade_estoque': [50]*10, 'fabricante': ['EMS']*10,
        'cidade': ['Recife']*10, 'nome_farmacia': ['Teste']*10
    })

df_remessas_base = distrib[['cod_farmacia', 'cod_medicamento', 'quantidade_estoque', 'fabricante']].copy()

# --- 2. LÓGICA DE QUANTIDADE ---

# Limpeza para cálculo
def get_clean_stock(val):
    try:
        v = float(val)
        return max(0, int(v))
    except:
        return 0

stock_clean = df_remessas_base['quantidade_estoque'].apply(get_clean_stock)
fator_aumento = np.random.uniform(1.20, 1.50, size=len(stock_clean))
total_semestre = stock_clean * fator_aumento

# Se o estoque for 0, definimos um envio mínimo padrão
total_semestre = np.where(total_semestre == 0, 60, total_semestre)

# Quantidade Fixa Mensal
df_remessas_base['qtd_mensal_fixa'] = np.ceil(total_semestre / 6).astype(int)

# --- 3. LÓGICA DE LABORATÓRIOS E LOTES ---

# A. Lista Oficial de Fabricantes)
laboratorios_oficiais = [
    "NewQuímica", "Medlove", "ENS", "Asiafarma",
    "Tenuto", "Pratik-Donaducci", "Axex", "Sandozan"
]

# Mapa Fixo: Nome -> ID
mapa_lab_id = {nome: 500 + i for i, nome in enumerate(laboratorios_oficiais)}

# Função para corrigir fabricantes nulos/sujos na base original antes de gerar ID
def corrigir_fabricante(val):
    # Se o valor não estiver na lista oficial (é nulo, sujo ou vazio), atribui um aleatório
    if val not in laboratorios_oficiais:
        return random.choice(laboratorios_oficiais)
    return val

# Cria uma coluna temporária limpa para garantir que todo mundo tenha ID
df_remessas_base['fabricante_corrigido'] = df_remessas_base['fabricante'].apply(corrigir_fabricante)

# Gera o ID baseando-se apenas nos nomes corrigidos
df_remessas_base['id_fabricante'] = df_remessas_base['fabricante_corrigido'].map(mapa_lab_id)

# Atualiza a coluna fabricante original para a corrigida
df_remessas_base['fabricante'] = df_remessas_base['fabricante_corrigido']

# B. Gerar Matriz de Lotes (Laboratório + Mês = Lote Único)
lotes_db = {}
meses = [7, 8, 9, 10, 11, 12]
ano = 2025

for lab_id in mapa_lab_id.values():
    for mes in meses:
        letras = "".join(random.choices("ABCDEF", k=2))
        nums = "".join(random.choices("0123456789", k=4))
        # Ex: EMS em Outubro sempre terá lote AB1234-10
        lotes_db[(lab_id, mes)] = f"{letras}{nums}-{mes}"

# --- 4. EXPANSÃO TEMPORAL ---

dfs_mensais = []

for mes in meses:
    # Copia a base calculada
    df_mes = df_remessas_base.copy()

    # 1. Define a Quantidade (é fixa mensal)
    df_mes['quantidade_recebida'] = df_mes['qtd_mensal_fixa']

    # 2. Define a Data de entrega do lote
    ultimo_dia = calendar.monthrange(ano, mes)[1]
    inicio_janela = ultimo_dia - 3

    # Função para gerar dia aleatório na janela de tempo que foi esoecificada
    def gerar_dia_entrega(row):
        dia = random.randint(inicio_janela, ultimo_dia)
        return f"{ano}-{mes:02d}-{dia:02d}"

    df_mes['data_entrega'] = df_mes.apply(gerar_dia_entrega, axis=1)

    # 3. Atribui o Lote
    df_mes['lote_fabricacao'] = df_mes['id_fabricante'].apply(lambda x: lotes_db.get((x, mes), "LOTE-GEN"))

    dfs_mensais.append(df_mes)

# Junta tudo
df_final_remessas = pd.concat(dfs_mensais, ignore_index=True)

# --- 5. AJUSTES FINAIS ---

# ID da Transação Único
df_final_remessas['id_transacao_remessa'] = range(1000000, 1000000 + len(df_final_remessas))

# Limpar colunas auxiliares que não vão para o CSV final
# Adicionado 'fabricante' na lista de colunas a manter
cols_to_keep = [
    'id_transacao_remessa', 'cod_farmacia', 'cod_medicamento',
    'fabricante', 'id_fabricante', # <--- Colunas Solicitadas
    'quantidade_recebida', 'data_entrega', 'lote_fabricacao'
]
df_final_remessas = df_final_remessas[cols_to_keep]


# --- 6. EXPORTAÇÃO ---

print(f"Histórico de Remessas (6 meses) gerado com {len(df_final_remessas)} linhas.")
print(df_final_remessas.head(10))

filename = "historico_remessas_2sem.csv"
df_final_remessas.to_csv(filename, index=False, encoding='utf-8')
print(f"Arquivo '{filename}' salvo com sucesso!")
```

## Tabela fato de vendas e tabela dimensão de clientes




```python
# --- CONFIGURAÇÕES ---
random.seed(42)
np.random.seed(42)

# Datas (2º Semestre 2025)
START_DATE = datetime(2025, 7, 1)
END_DATE = datetime(2025, 12, 31)
DAYS_RANGE = (END_DATE - START_DATE).days

# --- 1. CARREGAR DADOS ---
print("Carregando bases de dados...")
try:
    df_estoque = pd.read_csv("distribuicao_medicamentos_completa.csv")
    df_remessas = pd.read_csv("historico_remessas_2sem.csv")
    df_produtos = pd.read_csv("lista_medicamentos_expandida.csv")

    # Mapeamento de Preço e Categoria para consulta rápida
    prod_info = df_produtos.set_index('cod_medicamento')[['preco_unitario', 'categoria', 'substancia_principal']].to_dict('index')

except FileNotFoundError:
    print("ERRO: Arquivos anteriores não encontrados. Execute os passos anteriores.")
    # Mock para evitar quebra do script se rodado isoladamente
    exit()

# --- 2. CÁLCULO DE VENDAS NECESSÁRIAS (CONSISTÊNCIA MATEMÁTICA) ---
print("Calculando volume de vendas para garantir consistência...")

# A. Total Recebido por Farmácia/Produto
remessas_agg = df_remessas.groupby(['cod_farmacia', 'cod_medicamento'])['quantidade_recebida'].sum().reset_index()
remessas_agg.rename(columns={'quantidade_recebida': 'qtd_recebida_total'}, inplace=True)

# B. Estoque Atual
estoque_agg = df_estoque[['cod_farmacia', 'cod_medicamento', 'quantidade_estoque']].copy()

# C. Cruzamento (Merge)
df_calc = pd.merge(remessas_agg, estoque_agg, on=['cod_farmacia', 'cod_medicamento'], how='left')
df_calc['quantidade_estoque'] = df_calc['quantidade_estoque'].fillna(0)

# D. Lógica: O que entrou - O que sobrou = O que foi vendido
df_calc['qtd_vendida_total'] = df_calc['qtd_recebida_total'] - df_calc['quantidade_estoque']

# Ajuste de segurança (não pode haver vendas negativas, se houver erro no estoque, zeramos a venda)
df_calc['qtd_vendida_total'] = df_calc['qtd_vendida_total'].apply(lambda x: x if x > 0 else 0)

# Filtrar apenas o que tem venda para processar
items_to_sell = df_calc[df_calc['qtd_vendida_total'] > 0].copy()

print(f"Total de itens unitários a serem vendidos na simulação: {items_to_sell['qtd_vendida_total'].sum()}")

# --- 3. GERAÇÃO DA TABELA CLIENTES ---
print("Gerando base de clientes...")

NUM_CLIENTES_CADASTRADOS = 3533
clientes_data = []

nomes_base = ["Maria", "João", "Ana", "Pedro", "Lucas", "Mariana", "Carlos", "Fernanda", "Paulo", "Juliana", "Marcos", "Larissa", "Gabriel", "Camila", "Rafaela", "Patrícia", "Leandro", "Taís", "Roberto", "Maurício", "José", "Josiane", "Cleonice"]
sobrenomes = ["Silva", "Santos", "Oliveira", "Souza", "Rodrigues", "Ferreira", "Alves", "Pereira", "Lima", "Gomes", "Costa", "Ribeiro", "Barreto", "Marcondes", "Braga", "Sousa", "Bonfim", "Feitosa"]
dominios = ["gmail.com", "outlook.com", "uol.com.br", "hotmail.com", "bol.com.br"]

def gerar_cpf():
    return "".join([str(random.randint(0, 9)) for _ in range(11)])

for i in range(NUM_CLIENTES_CADASTRADOS):
    nome = f"{random.choice(nomes_base)} {random.choice(sobrenomes)}"
    cpf = gerar_cpf()
    email = f"{nome.lower().replace(' ', '.').replace('ç', 'c')}@{random.choice(dominios)}"

    clientes_data.append({
        "id_cliente": 10000 + i, # IDs 10000 a 12999 são cadastrados
        "nome_cliente": nome,
        "cpf": cpf,
        "email": email,
        "perfil_comprador": random.choice(["Uso Contínuo", "Geral", "Geral"])
    })

df_clientes = pd.DataFrame(clientes_data)

# Lista de IDs para sorteio
ids_cadastrados = df_clientes['id_cliente'].tolist()

# --- 4. MOTOR DE SIMULAÇÃO DE VENDAS ---
print("Simulando transações com lógica de mercado (Sazonalidade, Pareto e Meios de Pagamento)...")

transacoes_list = []
transacao_id_counter = 5000000
cliente_anonimo_counter = 9000000

# Agrupar itens por farmácia
grouped_lojas = items_to_sell.groupby('cod_farmacia')

# Range de 1 a 4 no total de compras por cliente, mas com peso maiores pesos para < 15
opcoes_range = list(range(1, 41)) # 1 a 40
pesos_range = []

for x in opcoes_range:
    if x <= 10:
        pesos_range.append(15.0)  # Altamente provável
    elif x <= 15:
        pesos_range.append(9.0)   # provável
    elif x <= 25:
        pesos_range.append(3.0)   # Médio
    elif x <= 30:
        pesos_range.append(0.9)   # Pouco provável
    elif x <= 35:
        pesos_range.append(0.5)   # Baixo
    else:
        pesos_range.append(0.1)   # Muito baixo

limites_compras_cliente = {
    cid: random.choices(opcoes_range, weights=pesos_range, k=1)[0]
    for cid in ids_cadastrados
}

contador_compras_cliente = {}

# --- PREPARAÇÃO DE PESOS ---

# 1. Pesos para Clientes Identificados (Lei de Pareto: 20% compram 80%)
# Isso evita que todos os clientes comprem a mesma quantidade
if ids_cadastrados:
    # O parâmetro a=1.16 aproxima a regra 80/20
    pesos_clientes = np.random.pareto(a=1.16, size=len(ids_cadastrados))
else:
    pesos_clientes = []

# 2. Pesos para Datas (Sazonalidade Mensal e Semanal)
pesos_datas = []
lista_datas_validas = []

for d in range(DAYS_RANGE + 1):
    curr_date = START_DATE + timedelta(days=d)
    day_num = curr_date.day
    weekday = curr_date.weekday() # 0=Seg, 6=Dom

    weight = 1.0

    # Efeito Payday (Dia 1 ao 10 vendem mais)
    if 1 <= day_num <= 10: weight = 1.8
    # Fim de mês (Dia 21+) vende menos
    elif day_num > 20: weight = 0.7

    # Efeito Fim de Semana (Sexta e Sabado fortes)
    if weekday == 4 or weekday == 5: weight *= 1.3

    # Efeito Sazonalidade Anual (Dezembro forte)
    if curr_date.month == 12: weight *= 1.4

    pesos_datas.append(weight)
    lista_datas_validas.append(curr_date)

# Normaliza para usar no random.choices
soma_pesos = sum(pesos_datas)
probs_datas = [p / soma_pesos for p in pesos_datas]

# Lista base de pagamentos
lista_metodos = ["Cartão de Crédito", "Cartão de Débito", "Dinheiro", "PIX", "Farmácia Popular", "Vale Alimentação"]

# --- LOOP DE VENDAS ---

ids_lojas = list(grouped_lojas.groups.keys())
lojas_outliers = set(random.sample(ids_lojas, k=int(len(ids_lojas) * 0.10)))

for cod_farmacia, group in grouped_lojas:

    if cod_farmacia in lojas_outliers:
        fator_loja = random.uniform(4.0, 6.0) # Vende 4x a 6x o esperado
    else:
        fator_loja = random.uniform(0.7, 1.3) # Variação normal

    # Monta pilha de produtos aplicando o fator
    pilha_produtos = []
    for _, row in group.iterrows():
        cod_med = row['cod_medicamento']
        qtd_base = int(row['qtd_vendida_total'])

        # AQUI APLICAMOS O FATOR:
        qtd_final = int(qtd_base * fator_loja)

        if qtd_final > 0:
            pilha_produtos.extend([cod_med] * qtd_final)

    # --- NOVA LÓGICA DE OUTLIERS ---
    # Se for uma loja outlier, duplicamos o estoque de venda para simular altíssima demanda
    # Isso fará com que essa loja tenha o dobro de transações/faturamento das outras
    if cod_farmacia in lojas_outliers:
        pilha_produtos = pilha_produtos * 2  # Multiplica volume de vendas por 2
        # Ou use: pilha_produtos.extend(random.sample(pilha_produtos, k=len(pilha_produtos)//2)) para 1.5x

    random.shuffle(pilha_produtos)

    while len(pilha_produtos) > 0:

        # 1. Definir Tamanho da Cesta (Cauda longa para outliers)
        # Maioria compra 1-3 itens, mas existem compras de abastecimento (10+)
        basket_size = random.choices(
            [1, 2, 3, 4, 5, 6, 7, 8, 10, 12, 15, 20],
            weights=[22, 20, 15, 10, 8, 6, 5, 4, 3, 2, 1, 0.2]
        )[0]

        if basket_size > len(pilha_produtos):
            basket_size = len(pilha_produtos)

        cesta = [pilha_produtos.pop() for _ in range(basket_size)]

        # 2. Calcular Valor Total
        valor_total = 0
        codigos_na_cesta = []
        contagem_cesta = {}

        for item in cesta:
            contagem_cesta[item] = contagem_cesta.get(item, 0) + 1

        for cod_med, qtd in contagem_cesta.items():
            info = prod_info.get(cod_med, {'preco_unitario': 10.0})
            valor_total += info['preco_unitario'] * qtd
            codigos_na_cesta.extend([str(cod_med)] * qtd)

        # 3. Definir Data (Com Sazonalidade)
        data_venda = random.choices(lista_datas_validas, weights=probs_datas, k=1)[0]

        # 4. Definir Cliente
        is_identificado = random.random() > 0.60
        id_cliente = cliente_anonimo_counter

        if is_identificado and ids_cadastrados:
            cand_id = random.choices(ids_cadastrados, weights=pesos_clientes, k=1)[0]

            # --- VALIDAÇÃO DO TETO ---
            feito = contador_compras_cliente.get(cand_id, 0)
            limite = limites_compras_cliente.get(cand_id, 50) # Pega o limite dele

            if feito < limite:
                id_cliente = cand_id
                contador_compras_cliente[cand_id] = feito + 1 # Incrementa contador
            else:
                # Se estourou o limite, vira cliente anônimo
                cliente_anonimo_counter += 1
        else:
            cliente_anonimo_counter += 1

        # 5. Definir Método de Pagamento (Baseado no Valor)
        # ["Crédito", "Débito", "Dinheiro", "PIX", "Farmácia Pop", "Vale"]

        if valor_total < 25.0:
            # Compras pequenas: Dinheiro e PIX fortes
            w_pag = [0.10, 0.25, 0.35, 0.25, 0.05, 0.0]
        elif valor_total < 120.0:
            # Compras médias: Débito e PIX dominam
            w_pag = [0.30, 0.35, 0.10, 0.20, 0.02, 0.03]
        else:
            # Compras grandes: Crédito domina absoluto
            w_pag = [0.65, 0.15, 0.05, 0.10, 0.0, 0.05]

        metodo = random.choices(lista_metodos, weights=w_pag, k=1)[0]

        # Registro
        transacoes_list.append({
            "id_transacao": transacao_id_counter,
            "cod_farmacia": cod_farmacia,
            "cod_medicamento": ",".join(codigos_na_cesta),
            "qtd_produtos": len(cesta),
            "data_venda": data_venda.strftime("%Y-%m-%d"),
            "valor_total_compra": round(valor_total, 2),
            "metodo_pagamento": metodo,
            "id_cliente": id_cliente
        })

        transacao_id_counter += 1

# --- 5. EXPORTAÇÃO ---

df_vendas = pd.DataFrame(transacoes_list) # Cria o DF

print("Aplicando sujeira nos dados...")

# Aplique a sujeira DIRETAMENTE neste df_vendas
# A. Nulos
mask_null = np.random.rand(len(df_vendas)) < 0.01
df_vendas.loc[mask_null, 'metodo_pagamento'] = np.nan

# B. Duplicatas
n_dup = int(len(df_vendas) * 0.005)
duplicatas = df_vendas.sample(n=n_dup)
df_vendas = pd.concat([df_vendas, duplicatas], ignore_index=True)

# C. Datas Inválidas/Futuras
# Escolhemos 5 indices aleatórios para colocar data errada
indices_erro = random.sample(range(len(df_vendas)), 5)
for idx in indices_erro:
    df_vendas.at[idx, 'data_venda'] = "2030-01-01"

# D. Valor Negativo (Estorno registrado errado)
idx_neg = random.randint(0, len(df_vendas)-1)
df_vendas.at[idx_neg, 'valor_total_compra'] = -50.00


# Tabela Vendas
print(f"Vendas geradas: {len(df_vendas)} transações.")
print("Exemplo de Venda:")
print(df_vendas.head())

filename_vendas = "tabela_vendas.csv"
df_vendas.to_csv(filename_vendas, index=False, encoding='utf-8')

# Tabela Clientes (Apenas os identificados para a dimensão)
# Os anonimos (ID > 9000000) não vão para a tabela dimensão, simulando que não tem cadastro
filename_clientes = "dim_clientes.csv"
# Removemos a coluna auxiliar de perfil antes de salvar
df_clientes.drop(columns=['perfil_comprador']).to_csv(filename_clientes, index=False, encoding='utf-8')

print(f"Arquivos '{filename_vendas}' e '{filename_clientes}' salvos com sucesso!")
```

## Criação tabela fato de ocorrências logísticas


```python
# --- CONFIGURAÇÕES ---
random.seed(42)
NUM_OCORRENCIAS = 5000 # Quantidade de problemas registrados no semestre

# --- 1. CARREGAR DADOS ---
try:
    # Precisamos saber o que tem em cada farmácia para gerar ocorrências reais
    df_estoque = pd.read_csv("distribuicao_medicamentos_completa.csv")

    # Para pegar o preço (para calcular prejuízo)
    df_produtos = pd.read_csv("lista_medicamentos_expandida.csv")
    dict_precos = df_produtos.set_index('cod_medicamento')['preco_unitario'].to_dict()

    print("Dados carregados para simulação de logística.")

except FileNotFoundError:
    print("ERRO: Arquivos base não encontrados.")
    # Mock
    exit()

# --- 2. DEFINIÇÃO DE TIPOS DE PROBLEMA ---
tipos_problema = [
    "Avaria - Transporte",       # 35%
    "Avaria - Temperatura",      # 10% (Focado em refrigerados)
    "Perda - Vencimento",        # 25% (Logística reversa falha)
    "Ajuste - Inventário (Perda)", # 20% (Furo de estoque)
    "Recusa de Recebimento"      # 10% (Falta de espaço)
]
pesos_problema = [0.35, 0.10, 0.25, 0.20, 0.10]

# --- 3. GERAR OCORRÊNCIAS ---

lista_ocorrencias = []
# Converte para lista de dicts para performance
estoque_sample = df_estoque.sample(n=NUM_OCORRENCIAS, replace=True).to_dict('records')

for i, row in enumerate(estoque_sample):

    categoria = row['categoria']
    tipo_escolhido = random.choices(tipos_problema, weights=pesos_problema)[0]

    # REGRAS DE NEGÓCIO (VIÉS)

    # Regra 1: Se for Refrigerado, chance de ser Temperatura aumenta drasticamente
    if categoria == "Refrigerados":
        if random.random() < 0.60: # 60% de chance de ser temperatura
            tipo_escolhido = "Avaria - Temperatura"

    # Regra 2: Se NÃO for refrigerado, não pode ser Temperatura
    elif tipo_escolhido == "Avaria - Temperatura":
        tipo_escolhido = "Avaria - Transporte" # Corrige para transporte

    # Regra 3: Controlados tem mais chance de 'Ajuste de Inventário' (Risco)
    if categoria == "Antibióticos e Controlados" and random.random() < 0.30:
        tipo_escolhido = "Ajuste - Inventário (Perda)"

    # Quantidade afetada (geralmente pequena, 1 a 5 unidades, exceto recusa de lote)
    qtd_afetada = random.randint(1, 5)
    if tipo_escolhido == "Recusa de Recebimento":
        qtd_afetada = random.randint(10, 50) # Recusa caixas inteiras

    # Calcular Prejuízo Estimado
    dict_custos = df_produtos.set_index('cod_medicamento')['preco_compra'].to_dict()

    # No loop de geração:
    custo = dict_custos.get(row['cod_medicamento'], 5.0)
    prejuizo = round(qtd_afetada * custo, 2) # Agora calcula sobre o custo real

    # Data aleatória no semestre
    start_date = datetime(2025, 7, 1)
    random_date = start_date + timedelta(days=random.randint(0, 180))

    lista_ocorrencias.append({
        "id_ocorrencia": 900000 + i,
        "cod_farmacia": row['cod_farmacia'],
        "cod_medicamento": row['cod_medicamento'],
        "categoria_produto": categoria, # Redundante, mas útil para analise rápida
        "data_ocorrencia": random_date.strftime("%Y-%m-%d"),
        "tipo_movimento": tipo_escolhido,
        "quantidade": qtd_afetada,
        "custo_prejuizo": prejuizo,
        "responsavel_registro": random.choice(["Sistema WMS", "Conferente", "Farmacêutico", "Auditoria"])
    })

# --- 4. EXPORTAR ---
df_logistica = pd.DataFrame(lista_ocorrencias)

print(f"Geradas {len(df_logistica)} ocorrências logísticas.")
print("Exemplo de Avaria de Temperatura:")
print(df_logistica[df_logistica['tipo_movimento'] == 'Avaria - Temperatura'].head(3))

filename = "ocorrencias_logisticas.csv"
df_logistica.to_csv(filename, index=False, encoding='utf-8')
print(f"Arquivo '{filename}' salvo com sucesso!")
```


## Criação tabela dimensão das capacidades de cada farmacia


```python
# --- CONFIGURAÇÕES ---
random.seed(42)

# --- 1. CARREGAR FARMÁCIAS ---
try:
    df_farmacias = pd.read_csv("dim_farmacias_final.csv")
    print(f"Carregando {len(df_farmacias)} farmácias.")
except FileNotFoundError:
    print("ERRO: 'dim_farmacias_final.csv' não encontrado.")
    # Mock
    df_farmacias = pd.DataFrame([{'cod_farmacia': 1, 'nome_farmacia': 'Teste', 'cidade': 'Recife', 'bairro': 'Boa Viagem'}])

# --- 2. MAPEAMENTO DE BAIRROS NOBRES (SHOPPING/PREMIUM) ---
# Lista baseada nos bairros gerados no script da dimensão farmácias
# Se a farmácia estiver aqui, assumimos perfil "Shopping Center" ou "Área Nobre"
bairros_premium_shopping = [
    "Boa Viagem", "Pina", "Casa Forte", "Espinheiro", # Recife
    "Barra", "Ondina", "Itaigara", "Caminho das Árvores", "Horto Florestal", # Salvador
    "Meireles", "Aldeota", "Papicu", "Dionísio Torres", # Fortaleza
    "Ponta Verde", "Jatiúca", "Pajuçara", # Maceió
    "Ponta Negra", "Tirol", "Petrópolis", # Natal
    "Tambaú", "Manaíra", "Cabo Branco", "Bessa", # João Pessoa
    "Renascença", "Ponta d'Areia", "Calhau", # São Luís
    "Jockey", "Fátima", "Ininga", # Teresina
    "Jardins", "13 de Julho", "Garcia", # Aracaju
    "Maurício de Nassau", # Caruaru (Interior Nobre)
    "Nova Betânia", # Mossoró
    "Catolé" # Campina Grande
]

# --- 3. FUNÇÃO DE DEFINIÇÃO DE PERFIL ---

def definir_perfil_logistico(row):
    bairro = str(row['bairro']).strip()
    cidade = row['cidade']

    # Regra 1: Centro é sempre Centro
    if "Centro" in bairro or "Comércio" in bairro or "Cidade Alta" in bairro:
        return "Loja de Rua (Centro)"

    # Regra 2: Bairros Nobres tendem a ser Shopping ou Premium
    elif bairro in bairros_premium_shopping:
        # 60% de chance de ser dentro de Shopping nestes bairros, 40% Loja de Rua Premium
        # Para simplificar categorias, vamos classificar como "Shopping Center"
        # (considerando que lojas de rua em áreas nobres têm custos similares a shopping)
        return "Shopping Center"

    # Regra 3: Aeroportos (Inserção forçada para variedade do dataset)
    # Vamos pegar aleatoriamente algumas farmácias para serem Aeroporto se não cair nas regras acima
    elif random.random() < 0.02 and cidade in ["Recife", "Salvador", "Fortaleza"]:
        return "Aeroporto/Rodoviária"

    # Regra 4: Todo o resto é Bairro Residencial
    else:
        return "Loja de Rua (Bairro)"

# --- 4. GERAÇÃO DOS DADOS ---

dados_capacidade = []

for _, row in df_farmacias.iterrows():

    perfil = definir_perfil_logistico(row)

    # Definição de Capacidades baseada no Perfil Real
    if perfil == "Aeroporto/Rodoviária":
        # Lojas micro, aluguel caríssimo, pouco estoque
        area = random.randint(30, 60)
        cap_geladeira = 50 # Litros (Frigobar clínico)
        cap_controlados = 100 # Unidades (Baixíssimo)
        custo_m2 = "Extremo"
        docas = 0 # Recebimento pela frente da loja

    elif perfil == "Shopping Center":
        # Lojas compactas, vitrine cara, estoque verticalizado
        area = random.randint(60, 120)
        cap_geladeira = 150
        cap_controlados = 600
        custo_m2 = "Alto"
        docas = 1 # Doca compartilhada do shopping

    elif perfil == "Loja de Rua (Centro)":
        # Lojas antigas, compridas, alto giro, estoque bagunçado
        area = random.randint(100, 200)
        cap_geladeira = 300 # Geladeira expositora dupla
        cap_controlados = 2000 # Cofre grande
        custo_m2 = "Médio"
        docas = 1 # Descarga na calçada ou fundos

    else: # Loja de Rua (Bairro)
        # Lojas amplas, terreno barato, servem como mini-hubs
        area = random.randint(150, 400)
        cap_geladeira = 400 # Várias geladeiras
        cap_controlados = 1500
        custo_m2 = "Baixo"
        docas = 2 # Estacionamento próprio facilita descarga

    # Variação aleatória realista (+- 10%)
    area = int(area * random.uniform(0.95, 1.05))
    cap_geladeira = int(cap_geladeira * random.uniform(0.9, 1.1))

    dados_capacidade.append({
        "cod_farmacia": row['cod_farmacia'],
        "nome_farmacia": row['nome_farmacia'],
        "cidade": row['cidade'],
        "bairro_localizacao": row['bairro'], # Mantemos para validação
        "perfil_loja": perfil,
        "area_total_m2": area,
        "capacidade_geladeira_litros": cap_geladeira,
        "capacidade_armario_controlados_un": cap_controlados,
        "qtd_docas_descarga": docas,
        "custo_operacional_m2": custo_m2
    })

# --- 5. EXPORTAR ---
df_cap = pd.DataFrame(dados_capacidade)

print("Amostra de Capacidades:")
print(df_cap[['nome_farmacia', 'perfil_loja', 'capacidade_geladeira_litros']].head())

filename = "dim_capacidade_farmacia.csv"
df_cap.to_csv(filename, index=False, encoding='utf-8')
print(f"Arquivo '{filename}' salvo com sucesso!")
```
